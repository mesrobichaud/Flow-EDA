## Import and Scaling

```python
# DEFINITIONS
def importdata(location, prefix, file_array): 
    FCS=[]
    for i in file_array:
        FCS.append(FlowCal.transform.to_rfi(FlowCal.io.FCSData(
                        str(location / str(prefix + '.' + format(i,'03'))))))
    return FCS

# this method takes in the Agglomerative clustering platelet data (i.e. platelet-only events)
# and generates an optimized ellipse (for gating) on the condition that
def plate_gate():
    return 0

#need to know what cells and the dataset
#assumes first datafile is the "control" as is done in manual gating 
def scaler(cells,labels,data):
    import scipy.cluster.hierarchy as sch
    from sklearn.cluster import AgglomerativeClustering
    
    if ((cells[0]=='Platelets') & (labels.iloc[0].System == 'WB')):
        test = np.column_stack((np.log(data[0][:,0]),np.log(data[0][:,1]),np.log(data[0][:,2])))
        clean = test[(test[:,2] > 5)]
        
        # create clusters
        hc = AgglomerativeClustering(n_clusters=2, affinity = 'euclidean', linkage = 'ward')
        if (len(clean[:,0]) < 40000):
            y_hc = hc.fit_predict(clean)
        else: raise Exception('Too many events to cluster, try raising the FL1 cut-off') 
        
        #delete population with larger size (RBCs and WBCs)  
        if (np.mean(clean[y_hc==0,0]) < np.mean(clean[y_hc==1,0])):
            clean = clean[y_hc ==0]
        else: clean = clean[y_hc == 1]
        
        #create an array of statistics about the population in FSC 
        #represents 3D away from mean, or 99.7% smaller than platelet events
        perc = np.percentile(clean[:,0],0.03) 
        
        #determine shift the platelet population such that PMPs are below FSC=1
        shift_0 = perc-1
        
        # TODO
        # - remove low CD61 events (how to determine threshold?)
        # - Define elliptical gating region in the "clean" population, 
        #.   return both the ungated and gated (linearized) populations
        
        #determine threshold of events to cut off based on what can be considered noise
        #initializing such that the 0th label is noise 
        gmm = sklearn.mixture.GaussianMixture(
                        n_components=2,covariance_type='tied', 
                        means_init=[[np.log(4)],[np.log(200)]]).fit_predict(np.log(data[0][:,2]).reshape(-1,1))
        #calculate cutoff
        cut_off = np.max(data[0][gmm==0,2])
        
        
        
        #transform each sample in data
        data_scaled = []
        data_s_g = []
        for sample in data: 
            # apply cutoff
            cut = sample[sample[:,2]>cut_off]
            # create new data files
            new = np.stack([np.log(cut[:,0])-shift_0,
                                                np.log(cut[:,1]),
                                                np.log(cut[:,2]),
                                                np.log(cut[:,3]),
                                                np.log(cut[:,4])],axis=1)
            data_scaled.append(new)
            data_s_g.append(1) # change 1 to plate_gate
        return [data_scaled,clean]
```

```python
def gate_function(pts):
    from sklearn.neighbors import KernelDensity
    kde = KernelDensity(kernel='gaussian', bandwidth=0.2).fit(pts.T)
    col =kde.score_samples(pts.T)
    new = pts[:,col.T > np.percentile(col.T,5)]
    
    pts = new

    #algorithm
    d = 2
    N = len(pts[0,:])

    #add a row of ones, so shape = 3xN
    Q = np.vstack([pts, np.ones(N)])

    #initialize
    count = 1
    err = 1
    u = (1/N) * np.ones([N,1])
    tolerance = 0.05

    #Khachiyan Algorithm
    while err > tolerance:
        X = np.linalg.multi_dot([Q,u*np.eye(N),Q.T])
        M = np.diag(np.linalg.multi_dot([Q.T,np.linalg.inv(X),Q]))
        maxm = np.amax(M)
        j = np.where(M == np.amax(M))[0]
        step_size = (maxm - d -1)/((d+1)*(maxm-1))
        new_u = (1 - step_size)*u
        new_u[j] +=  step_size
        err = np.linalg.norm(new_u - u)
        count = count + 1
        u = new_u
    U = u*np.eye(N)
    A = (1/d) * np.linalg.inv(np.linalg.multi_dot([pts,U,pts.T])
                              - np.linalg.multi_dot(
                                  [pts,u,np.transpose(pts @ u)] ))
    c = pts @ u
    eqn = V.T @ A @ V
    eqn2 = A[0,0]*V[0]**2 + 2*A[0,1]*V[0]*V[1]+A[1,1]*V[1]**2
    eqn2 = A[0,0]*(XX-c[0])**2 + 2*A[0,1]*(XX-c[0])*(YY-c[1])+A[1,1]*(YY-c[1])**2
    Z = np.ones([1000,1000])
    plt.subplot(1,1,1)
    plt.scatter(pts[0,:],pts[1,:],alpha=0.2)
    plt.plot(c[0],c[1],'s',c='r')
    plt.contour(XX,YY,eqn2,[1])
```
